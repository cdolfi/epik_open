```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
goemotion_bert = pd.read_csv('../data/goemotion_bert_volunteer_1000.csv', index_col=0)
goemotion_bert.head()
```

```python
eval_text = pd.read_csv('../data/evaluation_volunteer_text_1000.csv', index_col=0)
eval_text.head()
```

```python
data = []
for i in range(len(eval_text)):
    text = eval_text.iloc[i].values.item()
    indices = goemotion_bert.iloc[i].values > 0
    labels = goemotion_bert.columns.values[indices]
    scores = goemotion_bert.iloc[i].values[indices]
    data.append([text, labels, scores])
df = pd.DataFrame(data, columns=['text', 'labels', 'scores'])
df.head()
```

```python
df.to_csv('../data/feedback_goemotion_bert_volunteer_1000.csv')
```
